```python
import pandas as pd
import numpy as np
```

```python
source = pd.read_csv('datasets/others/xinyi-source-new.csv')
sample = pd.read_csv('datasets/others/xinyi-sample.csv')
inducer = pd.read_csv('datasets/others/xinyi-inducer.csv')
```

```python
source_clean = source.melt(id_vars='Unnamed: 0').dropna()
source_clean['location'] = source_clean['Unnamed: 0'] + source_clean['variable']
source_clean['variable'] = source_clean['variable'].astype(int)
source_clean = source_clean.sort_values(['Unnamed: 0', 'variable'])[['value', 'location']]
sample_loc = dict(zip(source_clean['value'], source_clean['location']))
sample_loc
```

```python
rev_sample_loc = dict(zip(source_clean['location'], source_clean['value']))
rev_sample_loc
```

```python
adder = 3
```

```python
ind_idx = adder + 1
sample_clean = sample.melt(id_vars='Unnamed: 0').dropna()
sample_clean['value'] = sample_clean['value'].apply(lambda x: int(x)+(adder*12) if x not in ['N', 'P', 'P2'] else x + '-' + str(ind_idx))
sample_clean['Source Well'] = 'sample' + sample_clean['value'].astype(str).str.split('.', expand = True)[0]
sample_clean['Source Well'] = sample_clean['Source Well'].map(sample_loc)
sample_clean['Destination Well'] = sample_clean['Unnamed: 0'] + sample_clean['variable'].astype(str)
sample_clean['Transfer Volume'] = 240
sample_clean = sample_clean[['Source Well', 'Destination Well', 'Transfer Volume']]
sample_clean = sample_clean.reset_index(drop=True)
sample_clean
```

```python
inducer_clean = inducer.melt(id_vars='Unnamed: 0').dropna()
inducer_clean['Source Well'] = 'inducer' + inducer_clean['value'].astype(str).str.split('.', expand = True)[0] + '-' + str(ind_idx)
inducer_clean['Source Well'] = inducer_clean['Source Well'].map(sample_loc)
inducer_clean['Destination Well'] = inducer_clean['Unnamed: 0'] + inducer_clean['variable'].astype(str)
inducer_clean['Transfer Volume'] = 40
inducer_clean = inducer_clean[['Source Well', 'Destination Well', 'Transfer Volume']]
inducer_clean = inducer_clean.reset_index(drop=True)
inducer_clean
```

```python
picklist = pd.concat([sample_clean, inducer_clean]).sort_index(kind='merge').reset_index(drop=True)
picklist.to_csv('datasets/others/picklist-0{}.csv'.format((adder+1)), index=False)
```

#### Generate dest layout

```python
picklist = pd.read_csv('datasets/others/picklist-0{}.csv'.format((adder+1)))
picklist['row'] = picklist['Destination Well'].str[0]
picklist['col'] = picklist['Destination Well'].str[1:]
picklist['value'] = picklist['Source Well'].map(rev_sample_loc)
picklist_agg = picklist[['row', 'col', 'value']].groupby(['row', 'col']).agg(lambda x: '_'.join(x)).reset_index()
picklist_agg['col'] = picklist_agg['col'].astype(int)
picklist_agg = picklist_agg.sort_values(['row', 'col'])
picklist_agg
```

```python
mask = picklist_agg['value'].duplicated(keep=False)
picklist_agg.loc[mask, 'value'] = picklist_agg.loc[mask, 'value'] + '_' + picklist_agg.groupby('value').cumcount().add(1).astype(str)
#picklist_agg['value'] = picklist_agg['value'].str[:-1] + '_' + picklist['value'].str[-1]
picklist_agg
```

```python
picklist_pvt = picklist_agg.pivot(index='row', columns='col', values='value').reset_index()#
picklist_pvt.fillna('empty', inplace=True)
picklist_pvt.loc[-1] = np.NaN
picklist_pvt.index = picklist_pvt.index + 1  # shifting index
picklist_pvt.sort_index(inplace=True)
#picklist_pvt.loc[-1] = np.NaN
picklist_pvt[24] = np.NaN
picklist_pvt = pd.concat([picklist_pvt['row'], pd.Series(['empty']*16), picklist_pvt.drop('row', axis=1)], axis=1)
picklist_pvt.fillna('empty', inplace=True)
picklist_pvt.iloc[0, 0] = 'A'
picklist_pvt.iloc[15, 0] = 'P'
picklist_pvt.to_csv('datasets/others/destination_layout_0{}.csv'.format((adder+1)), index=False)
```
